# TOTO予定データの取得

[001](001_データ収集.ipynb)/[002](002_データ加工.ipynb)は過去のデータの収集をしてきたが、予想の対象となるデータを見ていく。

TOTOは年毎に〇回目という通し番号を振りなおしておらず、URLで`holdCntId=X`でX回目の開催データが得られる。  
開催ではTOTO,Mini TOTO,TOTO Goalの3種があるが、必ずすべて開催されるわけではない。

```python
import pandas as pd

def _process_frame(tmp):
    return tmp[1:].drop(0,axis=1).drop(5,axis=1).rename(columns={1:tmp[1][0],2:tmp[2][0],3:tmp[3][0],4:"ホーム",6:"アウェイ"})

holdCntId = 200
url = "http://sport-kuji.toto-dream.com/dci/I/IPA/IPA01.do?op=disptotoLotInfo&holdCntId=%04d" % holdCntId
dfs = pd.io.html.read_html(url)
if len(dfs) is 20 or 24:
    print("ALL")
    tmp = dfs[3]
    if len(tmp) is 14:   # TOTO
        toto_df = _process_frame(tmp)
        print("TOTO")
        display(toto_df)
    tmp = dfs[8 + 1 if len(dfs) is 24 else 0]
    if len(tmp) is 6:    # Mini TOTO A
        minitoto_df_A = _process_frame(tmp)
        print("Mini TOTO")
        display(minitoto_df_A)
    tmp = dfs[13 + 2 if len(dfs) is 24 else 0]
    if len(tmp) is 6:   # Mini TOTO B
        minitoto_df_B = _process_frame(tmp)
        print("Mini TOTO")
        display(minitoto_df_B)
    tmp = dfs[18 + 3 if len(dfs) is 24 else 0]
    if len(tmp) is 4:   # TOTO GOAL
        totogoal_df = _process_frame(tmp)
        print("TOTO GOAL")
        display(totogoal_df)
elif len(dfs) is 18:       # Mini TOTO(A,B) & TOTO GOAL
    print("mini2 + goal")
    if len(dfs[3]) is 6:    # Mini TOTO A
        minitoto_df_A = _process_frame(tmp)
        print("Mini TOTO")
        display(minitoto_df_A)
    if len(dfs[9]) is 6:   # Mini TOTO B
        minitoto_df_B = _process_frame(tmp)
        print("Mini TOTO")
        display(minitoto_df_B)
    if len(dfs[15]) is 4:   # TOTO GOAL
        totogoal_df = _process_frame(tmp)
        print("TOTO GOAL")
        display(totogoal_df)
elif len(dfs) is 12:       # Mini TOTO & TOTO GOAL
    print("mini + goal")
    if len(dfs[3]) is 6:    # Mini TOTO A
        minitoto_df_A = _process_frame(tmp)
        print("Mini TOTO")
        display(minitoto_df_A)
    if len(dfs[9]) is 4:   # TOTO GOAL
        totogoal_df = _process_frame(tmp)
        print("TOTO GOAL")
        display(totogoal_df)    
elif len(dfs) is 6:   # TOTO GOALのみの場合
    print("goal")
    if len(dfs[3]) is 4:
        totogoal_df = _process_frame(tmp)
        print("TOTO GOAL")
        display(totogoal_df)
```

一通り取れそうなので、これを関数化した。

```python
def get_toto_schedule(holdCntId):
    """
    TOTOの対象ゲームを収集
    引数として開催回を指定する
    """
    toto = None
    miniA = None
    miniB = None
    goal = None
    url = "http://sport-kuji.toto-dream.com/dci/I/IPA/IPA01.do?op=disptotoLotInfo&holdCntId=%04d" % holdCntId
    dfs = pd.io.html.read_html(url)
    for df in dfs:
        if len(df)>=4 and df[1][0]=="開催日":
            if len(df)==14:
                toto = _process_frame(df)
            elif len(df) == 6:
                if miniA is None:
                    miniA = _process_frame(df)
                else:
                    miniB = _process_frame(df)
            elif len(df) == 4:
                goal = _process_frame(df)
    return toto, miniA, miniB, goal
```

## TOTOの結果を取得する

既に結果が出た開催データに関して、結果も公表されているので取得しておく

```python
url = "http://sport-kuji.toto-dream.com/dci/I/IPB/IPB01.do?op=lnkHoldCntLotResultLsttoto&holdCntId=" + str(1041)
dfs = pd.io.html.read_html(url)
if len(dfs) > 2:
    for i in range(len(dfs)):
        if dfs[i][1][0] == "競技場":
            print("result:",i)
            display(dfs[i])
        if dfs[i][1][0] == "1等":
            print("prize:",i)
            display(dfs[i])
else:
    print("Not held")
```

```python
def _process_result(tmp,hldCnt):
    if len(tmp.columns)==7: # toto/miniのみ。goalは除外
        tmp0 = tmp[2:].rename(columns={0:tmp[0][0],1:tmp[1][0],2:"index",3:tmp[1][1],4:tmp[2][1],5:tmp[3][1],6:tmp[4][1]}).reset_index(drop=True)
        tmp0["開催回"] = hldCnt
        return tmp0
    elif len(tmp.columns)==6:
        return None   # Goalは除外
    else:
        return None
    
_process_result(dfs[2],1041)
```

当選金に関しては、当選口数や繰越金はあまり重要なデータではないので、

第何回:当選金額

という情報に置き換える。

```python
import re
def _process_prize_frame(tmp):
    hldcnt = re.sub(r"\D","",tmp[0][0])   # 第〇回から数値だけ抽出
    if len(tmp.columns)==4: # TOTO
        tmp0 = tmp[1:].rename(columns={0:"開催回",1:tmp[1][0],2:tmp[2][0],3:tmp[3][0]})
        tmp0["開催回"].iloc[0] = hldcnt
        return tmp0.iloc[0:1].set_index("開催回")
    elif len(tmp.columns)==2: # mini TOTO
        tmp0 = tmp[1:].rename(columns={0:"開催回",1:tmp[1][0]})
        tmp0["開催回"].iloc[0] = hldcnt
        return tmp0.iloc[0:1].set_index("開催回")
    elif len(tmp.columns)==3: # toto GOAL
        tmp0 = tmp[1:].rename(columns={0:"開催回",1:tmp[1][0],2:tmp[2][0]})
        tmp0["開催回"].iloc[0] = hldcnt
        return tmp0.iloc[0:1].set_index("開催回")

_process_prize_frame(dfs[9])
```

```python
def get_toto_expection(holdCntId):
    url = "http://sport-kuji.toto-dream.com/dci/I/IPC/IPC01.do?op=initVoteRate&commodityId=01&holdCntId=%04d" % holdCntId
    dfs = pd.io.html.read_html(url)
    if len(dfs) > 2:
        for i in range(len(dfs)):
            if dfs[i][0][0] == "開催日":
                return dfs[i].dropna().rename(columns={4:"予想1",5:"予想0",6:"予想2"}).reset_index()
    return None

def get_miniA_expection(holdCntId):
    url = "http://sport-kuji.toto-dream.com/dci/I/IPC/IPC01.do?op=initVoteRate&commodityId=06&holdCntId=%04d" % holdCntId
    dfs = pd.io.html.read_html(url)
    if len(dfs) > 2:
        for i in range(len(dfs)):
            if dfs[i][0][0] == "開催日":
                return dfs[i].dropna().rename(columns={4:"予想1",5:"予想0",6:"予想2"}).reset_index()
    return None

def get_miniB_expection(holdCntId):
    url = "http://sport-kuji.toto-dream.com/dci/I/IPC/IPC01.do?op=initVoteRate&commodityId=07&holdCntId=%04d" % holdCntId
    dfs = pd.io.html.read_html(url)
    if len(dfs) > 2:
        for i in range(len(dfs)):
            if dfs[i][0][0] == "開催日":
                return dfs[i].dropna().rename(columns={4:"予想1",5:"予想0",6:"予想2"}).reset_index()
    return None


def get_goal_expection(holdCntId):
    url = "http://sport-kuji.toto-dream.com/dci/I/IPC/IPC01.do?op=initVoteRate&commodityId=02&holdCntId=%04d" %holdCntId
    dfs = pd.io.html.read_html(url)
    if len(dfs) > 2:
        for i in range(len(dfs)):
            if dfs[i][0][0] == "開催日":
                return dfs[i].dropna().rename(columns={5:"予想0点",6:"予想1点",7:"予想2点",8:"予想3点以上"}).reset_index()
    return None

get_toto_expection(1041)
```

```python
def get_toto_result(holdCntId):
    """
    TOTOの対象ゲームを収集
    引数として開催回を指定する
    """
    toto = None
    toto_prize = None
    miniA = None
    miniA_prize = None
    miniB = None
    miniB_prize = None
    goal = None
    goal_prize = None
    url = "http://sport-kuji.toto-dream.com/dci/I/IPB/IPB01.do?op=lnkHoldCntLotResultLsttoto&holdCntId=%04d" % holdCntId
    dfs = pd.io.html.read_html(url)
    for df in dfs:
        if len(df)>=4 and df[1][0]=="競技場":
            if len(df)==15:
                toto = _process_result(df,holdCntId)
                toto_exp = get_toto_expection(holdCntId)
                if toto_exp is not None:
                    toto["予想1"] = toto_exp["予想1"]
                    toto["予想0"] = toto_exp["予想0"]
                    toto["予想2"] = toto_exp["予想2"]
            elif len(df) == 7:
                if miniA is None:
                    miniA = _process_result(df,holdCntId)
                    miniA_exp = get_miniA_expection(holdCntId)
                    if miniA_exp is not None:
                        miniA["予想1"] = miniA_exp["予想1"]
                        miniA["予想0"] = miniA_exp["予想0"]
                        miniA["予想2"] = miniA_exp["予想2"]
                elif miniB is None:
                    miniB = _process_result(df,holdCntId)
                    miniB_exp = get_miniB_expection(holdCntId)
                    if miniB_exp is not None:
                        miniB["予想1"] = miniB_exp["予想1"]
                        miniB["予想0"] = miniB_exp["予想0"]
                        miniB["予想2"] = miniB_exp["予想2"]
                elif goal is None:
                    goal = _process_result(df,holdCntId)
                    goal_exp = get_goal_expection(holdCntId)
                    if goal is not None and goal_exp is not None:
                        goal["予想0点"] = goal_exp["予想0点"]
                        goal["予想1点"] = goal_exp["予想1点"]
                        goal["予想2点"] = goal_exp["予想2点"]
                        goal["予想3点以上"] = goal_exp["予想3点以上"]
        if len(df)>=3 and df[1][0]=="1等":
            if len(df.columns) == 4:
                toto_prize = _process_prize_frame(df)
            elif len(df.columns) == 2:
                if miniA_prize is None:
                    miniA_prize = _process_prize_frame(df)
                else:
                    miniB_prize = _process_prize_frame(df)
            elif len(df.columns) == 3:
                goal_prize = _process_prize_frame(df)
#    return toto, miniA, miniB, goal, toto_prize, miniA_prize, miniB_prize, goal_prize
    return toto, miniA, miniB, toto_prize, miniA_prize, miniB_prize
```
