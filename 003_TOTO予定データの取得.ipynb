# TOTO予定データの取得

[001](001_データ収集.ipynb)/[002](002_データ加工.ipynb)は過去のデータの収集をしてきたが、予想の対象となるデータを見ていく。

TOTOは年毎に〇回目という通し番号を振りなおしておらず、URLで`holdCntId=X`でX回目の開催データが得られる。  
開催ではTOTO,Mini TOTO,TOTO Goalの3種があるが、必ずすべて開催されるわけではない。

```python
import pandas as pd

def _process_frame(tmp):
    return tmp[1:].drop(0,axis=1).drop(5,axis=1).rename(columns={1:tmp[1][0],2:tmp[2][0],3:tmp[3][0],4:"ホーム",6:"アウェイ"})

holdCntId = 200
url = "http://sport-kuji.toto-dream.com/dci/I/IPA/IPA01.do?op=disptotoLotInfo&holdCntId=%04d" % holdCntId
dfs = pd.io.html.read_html(url)
if len(dfs) is 20 or 24:
    print("ALL")
    tmp = dfs[3]
    if len(tmp) is 14:   # TOTO
        toto_df = _process_frame(tmp)
        print("TOTO")
        display(toto_df)
    tmp = dfs[8 + 1 if len(dfs) is 24 else 0]
    if len(tmp) is 6:    # Mini TOTO A
        minitoto_df_A = _process_frame(tmp)
        print("Mini TOTO")
        display(minitoto_df_A)
    tmp = dfs[13 + 2 if len(dfs) is 24 else 0]
    if len(tmp) is 6:   # Mini TOTO B
        minitoto_df_B = _process_frame(tmp)
        print("Mini TOTO")
        display(minitoto_df_B)
    tmp = dfs[18 + 3 if len(dfs) is 24 else 0]
    if len(tmp) is 4:   # TOTO GOAL
        totogoal_df = _process_frame(tmp)
        print("TOTO GOAL")
        display(totogoal_df)
elif len(dfs) is 18:       # Mini TOTO(A,B) & TOTO GOAL
    print("mini2 + goal")
    if len(dfs[3]) is 6:    # Mini TOTO A
        minitoto_df_A = _process_frame(tmp)
        print("Mini TOTO")
        display(minitoto_df_A)
    if len(dfs[9]) is 6:   # Mini TOTO B
        minitoto_df_B = _process_frame(tmp)
        print("Mini TOTO")
        display(minitoto_df_B)
    if len(dfs[15]) is 4:   # TOTO GOAL
        totogoal_df = _process_frame(tmp)
        print("TOTO GOAL")
        display(totogoal_df)
elif len(dfs) is 12:       # Mini TOTO & TOTO GOAL
    print("mini + goal")
    if len(dfs[3]) is 6:    # Mini TOTO A
        minitoto_df_A = _process_frame(tmp)
        print("Mini TOTO")
        display(minitoto_df_A)
    if len(dfs[9]) is 4:   # TOTO GOAL
        totogoal_df = _process_frame(tmp)
        print("TOTO GOAL")
        display(totogoal_df)    
elif len(dfs) is 6:   # TOTO GOALのみの場合
    print("goal")
    if len(dfs[3]) is 4:
        totogoal_df = _process_frame(tmp)
        print("TOTO GOAL")
        display(totogoal_df)
```

無関係なフレームが多いが、とりたいデータは開催日/試合開始予定時間/競技場/ホーム/アウェイのデータである。

```python
holdCntId = 1043
url = "http://sport-kuji.toto-dream.com/dci/I/IPA/IPA01.do?op=disptotoLotInfo&holdCntId=%04d" % holdCntId
dfs = pd.io.html.read_html(url)
```

```python
for df in dfs:
    if df[1][0]=="開催日":
        if len(df)==14:
            print("TOTO")
            display(df)
        elif len(df) == 6:
            print("Mini TOTO")
            display(df)
        elif len(df) == 4:
            print("TOTO Goal")
            display(df)
```

一通り取れそうなので、これを関数化した。

```python
def get_toto_schedule(holdCntId):
    """
    TOTOの対象ゲームを収集
    引数として開催回を指定する
    """
    toto = None
    miniA = None
    miniB = None
    goal = None
    url = "http://sport-kuji.toto-dream.com/dci/I/IPA/IPA01.do?op=disptotoLotInfo&holdCntId=%04d" % holdCntId
    dfs = pd.io.html.read_html(url)
    for df in dfs:
        if len(df)>=4 and df[1][0]=="開催日":
            if len(df)==14:
                toto = _process_frame(df)
            elif len(df) == 6:
                if miniA is None:
                    miniA = _process_frame(df)
                else:
                    miniB = _process_frame(df)
            elif len(df) == 4:
                goal = _process_frame(df)
    return toto, miniA, miniB, goal
```

```python
toto, miniA, miniB, goal = get_toto_schedule(1043)
#display(miniB)
```

```python
from tqdm import tqdm
data = pd.DataFrame()
for i in tqdm(range(900,1041)):
    toto, miniA, miniB, goal = get_toto_schedule(i)
    if toto is not None:
        data = pd.concat([data, toto])
    if miniA is not None:
        data = pd.concat([data, miniA])
    if miniB is not None:
        data = pd.concat([data, miniB])
    if goal is not None:
        data = pd.concat([data, goal])
```
