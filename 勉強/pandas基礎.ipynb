## 注意点

データフレームを操作は、基本的に破壊的な操作はしないようになっています。

例えば、列名を変更する操作として、

````python
df.rename(columns={"年度":"year"})
````

でできますが、再度dfを見てみると、列名は変わっていません。

````python
display(df)
````

変更するには、
``` python
df = df.rename(columns={"年度":"year"})
```
としなければならないようになっています。

これは、ちょっとしたお試しをしてみたいときに、元のデータを破壊しないようになっているためと思われますが、個人的にはdf.renameってやった時点で変わっているつもりだったので「あれ？」となりました。

# numpyについて

pandasの前にnumpyを理解しておいた方が良いのでnumpyも調べておきます。

numpyは、Pythonの行列計算用ライブラリです。  
Matlab(すべての変数をdoubleの行列として扱う)と似ていて、アクセスの仕方もMatlabに似せていると思われます。

````python
import numpy as np # as npとしてimportするのが慣例らしい
ar = np.array([1,2,3])
print(ar[0])
````

このライブラリがなぜ重要かというと、Pythonは型宣言のないインタープリタ言語であるため、**for文の計算が非常に遅い**ためです。  
もともとインタープリタはfor文が遅いというところがありますが、numpyを使うと配列要素すべてに対して計算を実行することができるので、計算が速くなります。

````python
import numpy as np
a = np.ones(100000)
b=np.ones(100000 )

def dot(a, b):
    s = 0
    for i in range(len(a)):
        s += a[i] * b[i]
    return s
````

````python
%timeit dot(a, b)
````

````python
%timeit np.dot(a, b)
````

同じ計算をfor文で回して計算すると、かなりの差があることがわかります。  
実際には、[参考](https://atsuoishimoto.hatenablog.com/entry/2018/01/06/195649)を見る限り、for文のせいだけではなく、型宣言していないためいちいち変換していることが遅い要因だそうです。

とはいえ、計算が速いに越したことはないので、多くのライブラリがnumpyをベースにしています。
